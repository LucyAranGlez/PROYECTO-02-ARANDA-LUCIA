#Pt3 - statistics / estadistica

```python
print("media cantidad:", df['Quantity'].mean()) #media
print("media precio de unidad:", df['UnitPrice'].mean()) 
print("media id:", df['CustomerID'].mean())
```

```python
print("maximo cantidad:", df['Quantity'].max()) #maximo
print("maximo precio de unidad:", df['UnitPrice'].max()) 
print("maximo id:", df['CustomerID'].max())
```

```python
print("minimo cantidad:", df['Quantity'].min()) #minimo
print("minimo precio de unidad:", df['UnitPrice'].min()) 
print("minimo id:", df['CustomerID'].min())
```

```python
print("desviacion estandar cantidad:", df['Quantity'].std()) #desviacion estandar
print("desviacion estandar de unidad:", df['UnitPrice'].std()) 
print("desviacion estandar id:", df['CustomerID'].std())
```

#Pt4 - no sense values / valores sin sentido

```python
df = df[df.Quantity >= 0] #defining the condition / definir la condicion
df.describe() #result / resultado
```

#Pt5 - unique values / valores unicos

```python
print("valores unicos de cantidad:", df['Quantity'].unique().sum()) #valores unicos (ya que no especifica si num o no num, son todas las columnas)
print("valores unicos de pais:", df['Country'].unique().sum())
print("valores unicos de clienteid:", df['CustomerID'].unique().sum())
print("valores unicos de precio de unidad:", df['UnitPrice'].unique().sum())
print("valores unicos de fecha de factura:", df['InvoiceDate'].unique().sum())
print("valores unicos de descripcion:", df['Description'].unique().sum())
print("valores unicos de codigo de stock:", df['StockCode'].unique().sum())
print("valores unicos de codigo de no. de factura:", df['InvoiceNo'].unique().sum())
```
